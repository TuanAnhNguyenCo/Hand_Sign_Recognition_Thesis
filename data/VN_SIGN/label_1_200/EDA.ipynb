```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 1->200 total

```python
labels = pd.read_csv("full_data_1_200_three_view_ord1.csv")
label_counts = labels.loc[labels['label'] < 200]['label'].value_counts().sort_index()
num_subplots = 6
data_chunks = np.array_split(label_counts, num_subplots)
fig, axs = plt.subplots(3,2, figsize=(25,  20))
for i, data_chunk in enumerate(data_chunks):
    row = i // 2
    col = i % 2
    # Vẽ biểu đồ cột
    data_chunk.plot(kind='bar', title=f'Biểu đồ {i + 1}',ax=axs[row, col])
    axs[row, col].set_xlabel('Label')
    axs[row, col].set_ylabel('Count')
    
```
